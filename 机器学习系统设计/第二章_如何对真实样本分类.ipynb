使用 Iris 数据集进行单分类

```
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import numpy as np
import seaborn as sns

sns.set()
```

```
df = sns.load_dataset("iris")
print(df.head())
sns.pairplot(df, hue="species")
```

#### 创建第一个分类模型

从上图可以看出采用 petal_length 属性就可以分出 setosa 了。

setosa 的 petal_length 最大 1.9，非 setosa 的 petal_length 最小 3.0。所以可以简单的认为 petal_length 小于2.0 的都是 setosa。

**这个2.0就可以认为是某个维度上的一个简单阈值。**

```
df.groupby('species')['petal_length'].agg(['min','max'])
```
